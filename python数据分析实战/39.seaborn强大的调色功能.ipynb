```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
def sin_plot():
    x = np.linspace(0, 14, 100)
    plt.figure(figsize=(8,6))
    for i in range(4):
        plt.plot(x, np.sin(x+i)*(i+0.75), label='sin(x+%s)*(%s+0,75)' % (i, i))
    plt.legend()
```

```python
sin_plot()
```
