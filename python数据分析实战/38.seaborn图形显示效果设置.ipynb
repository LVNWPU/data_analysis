# axes style and set_style

```python
x = np.linspace(0, 14, 100)
```

```python
y1 = np.sin(x)
y2 = np.sin(x+2)*1.25
```

```python
def sin_plot():
    plt.plot(x, y1)
    plt.plot(x, y2)
```

```python
sin_plot()
```

```python
import seaborn as sns
```

```python
sin_plot()
```

```python
style = ['darkgrid', 'dark', 'white', 'whitegrid', 'tricks']
```

```python
sns.set_style(style[0], {'grid.color': 'red'})
```

```python
sin_plot()
```

```python
sns.axes_style()
```

```python
sin_plot()
```

```python
sns.set()  # 恢复默认值
```

```python
sin_plot()
```

```python
sns.set_style('white')
```

```python
sin_plot()
```

```python
sns.set()
```

```python
sin_plot()
```
