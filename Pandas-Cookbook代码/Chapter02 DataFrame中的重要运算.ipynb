# 串联DataFrame方法

```python
movie = pd.read_csv('data/movie.csv')
movie.isnull().head()  # 使用isnull方法将每个值转变为布尔值
```

```python
movie.isnull().sum()  # 使用sum统计布尔值，返回的是Series
```

```python
movie.isnull().sum().sum()  # 对这个Series再使用sum，返回整个DataFrame的缺失值的个数，返回值是个标量
```

```python
movie.any().any()   # 判断整个DataFrame有没有缺失值，方法是连着使用两个any
```

```python
movie.isnull().get_dtype_counts()  # isnull返回同样大小的DataFrame，但所有的值变为布尔值
```

```python
# movie数据集的对象数据包含缺失值。默认条件下，聚合方法min、max、sum，不会返回任何值
movie[['color', 'movie_title', 'color']].max()   
```

```python
movie.select_dtypes(['object']).fillna('').max()  # 选择出列的类型是object,并用''填充缺失值，选择出最大值
```

# 对比缺失值

```python
# Pandas使用NumPy NaN（np.nan）对象表示缺失值。这是一个不等于自身的特殊对象：
np.nan == np.nan
```

```python
# Python的None对象是等于自身的
None == None
```

```python
# 所有和np.nan的比较都返回False，除了不等于：
5 > np.nan
```

```python
np.nan > 5
```

```python
5 != np.nan
```

```python
# college_ugds_所有值和.0019比较，返回布尔值DataFrame
college = pd.read_csv('data/college.csv', index_col='INSTNM')
college_ugds_ = college.filter(like='UGDS_')
```

```python
college_ugds_.head() == .0019
```

```python
# 用DataFrame和DataFrame进行比较
college_self_compare = college_ugds_ == college_ugds_
college_self_compare
```

```python
# 用all()检查是否所有的值都是True；这是因为缺失值不互相等于
college_self_compare.all() 
```

```python
# 可以用==号判断，然后求和
(college_ugds_ == np.nan).sum()
```

```python
# 统计缺失值最主要方法是使用isnull方法
college_ugds_.isnull().sum()
```

```python
from pandas.testing import assert_frame_equal
```

```python
# 比较两个DataFrame最直接的方法是使用equals()方法
assert_frame_equal(college_ugds_, college_ugds_)
```

```python
college_ugds_.eq(.0019).head()
```

# 对DataFrame进行转置

```python
college = pd.read_csv('data/college.csv', index_col='INSTNM')
college_ugds_ = college.filter(like='UGDS_')
college_ugds_.head()
```

```python
college_ugds_.shape
```

```python
college_ugds_.count()  # 默认按行方向进行操作，count()返回非缺失值的个数
```

```python
college_ugds_.count(axis=0)
```

```python
college_ugds_.count(axis='index')
```

```python
college_ugds_.count(axis='columns').head()  # 按列方向，统计每行的非缺失值个数
```

```python
college_ugds_.sum(axis='columns').head()  # 按列方向求和
```

```python
college_ugds_.median(axis='index')  #  按行方向求中位数
```

```python
college_ugds_cumsum = college_ugds_.cumsum(axis=1)  # 按列方向求累积和
college_ugds_cumsum.head()
```

```python
college_ugds_cumsum.sort_values('UGDS_HISP', ascending=False) # 按某一列进行降序排列
```

# 实战：决定大学校园的多样性

```python
pd.read_csv('data/college_diversity.csv', index_col='School')
```

```python
college = pd.read_csv('data/college.csv', index_col='INSTNM')
college_ugds_ = college.filter(like='UGDS_')
college_ugds_.head()
```

```python
college_ugds_.isnull().sum(axis=1).sort_values(ascending=False).head()
```

```python
college_ugds_ = college_ugds_.dropna(how='all')  # 丢弃缺失值
college_ugds_.head()
```

```python
college_ugds_.isnull().sum()
```

```python
college_ugds_.ge(.15).head()  # 大于等于0.15
```

```python
diversity_metric = college_ugds_.ge(.15).sum(axis='columns')
diversity_metric.head()
```

```python
diversity_metric.value_counts()
```

```python
diversity_metric.sort_values(ascending=False).head()
```

```python
# 选出两行
college_ugds_.loc[['Regency Beauty Institute-Austin', 'Central Texas Beauty College-Temple']]
```
